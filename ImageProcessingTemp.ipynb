**Concatenate Function**

```python
def get_concat (im1, im2):
  dst = Image.new('RGB', (im1.width + im2.width, im1.height))
  dst.paste(im1, (0, 0))
  dst.paste(im2, (im1.width, 0))
  return dst
```

# **Color Representation of Images**

**Gray-Scale Image**

```python
from PIL import ImageOps
img_gray = ImageOps.grayscale(resize_image)
```

```python
img_gray.mode
```

```python
plt.imshow(img_gray)
```

**Quantization Of Image**

```python
for n in range(3,8):
  f = plt.figure()
  f.set_figwidth(5)
  f.set_figheight(5)
  plt.imshow(get_concat(img_gray,  img_gray.quantize(256//2**n))) 
  plt.title("256 Quantization Levels  left vs {}  Quantization Levels right".format(256//2**n))
  plt.show()
```
